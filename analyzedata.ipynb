# Analysis of Data 

```python
import matplotlib
```

```python
import pandas as pd
import numpy as np
import matplotlib 
```

```python
def get_data(size = 15):
  df = pd.DataFrame()
  df['Place'] = np.random.choice(['Walt Disney World Resort', 'Mall of America', 'Time Square', 'Las Vegas Strip', 'National Mall and Memorial Parks', 'Disneyland Resort'], size)
  df['Location'] = np.random.randint(0, 100, size)
  df['Visitors_millions'] = np.random.randint(0, 100, size)
  return df
```

```python
get_data().to_csv('get_data.csv')
df = pd.read_csv('get_data.csv')
df
```

```python
del df['Unnamed: 0']
df
```

```python
df.plot(title = 'Favorite-Vacation-Locations', xlabel = "Location", rot = 45, ylabel = "Visitors_million")
```

```python
# Plot a bar graph to show location to visitor ratio.
df.plot.bar(title = 'Favorite-Vacation-Locations', xlabel = "Locations", rot = 45, ylabel = "Visitors_million")
```

```python
# Plot a pie chart to show the first 14 Locations.
df['Location'].plot.pie()
```
